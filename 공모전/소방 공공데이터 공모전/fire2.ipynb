```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import random
from datetime import datetime

path = '/Windows/Fonts/Hancom Gothic Regular.ttf'
font_name = fm.FontProperties(fname=path, size=50).get_name()
plt.rc('font', family=font_name)

%matplotlib inline
mpl.rcParams['axes.unicode_minus'] = False

file_path = 'C:/fire/소방청_화재발생 주소정보_20171231.csv' # 데이터 1,000개만 확인
fire = pd.read_csv(file_path, sep=',', encoding='euc-kr')

# 데이터프레임 [시도, 시군구, 읍면동, 출동소요시간, 화재진압시간]
df_fire = pd.DataFrame(fire)
df = df_fire[['시도', '시군구', '읍면동', '출동소요시간', '화재진압시간']]

# 화재진압시간 정렬 / Top15
extinguish_time = df.sort_values('화재진압시간', ascending=False).head(15)

# 출동소요시간 정렬 / Top15
moveout_time = df.sort_values('출동소요시간', ascending=False).head(15)

display(extinguish_time)
display(moveout_time)
```
