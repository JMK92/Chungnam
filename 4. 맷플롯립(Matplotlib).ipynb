```python
# 차트제목, 축(x, y)제목

plt.title('Basic Flow Chart')
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel(['X-Label'])
plt.ylabel(['Y-Label'])
plt.show()
```

```python
# 축범위 :  axis()
plt.title('Basic Flow Chart')
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel(['X-Label'])
plt.ylabel(['Y-Label'])
# 축범위 axis() : [xmin, xmax, ymin, ymax]
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
plt.title('Basic Flow Chart')
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'k^') # 'bo', 'g-', '--'
plt.xlabel(['X-Label'])
plt.ylabel(['Y-Label'])
# 축범위 axis() : [xmin, xmax, ymin, ymax]
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
plt.title('Basic Flow Chart')      # 색이름으로 가능 # 'springgreen', 'violet'
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], color= '#e35f62', marker = 'o', linestyle = '--') 
plt.xlabel(['X-Label'])
plt.ylabel(['Y-Label'])
# 축범위 axis() : [xmin, xmax, ymin, ymax]
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
# 특정 영역에 색상으로 채워서 강조
# fill() - 채워라
# fill_between() : x축, y축 담아서
# fill_betweenx() : 위에랑 똑같음

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.fill_between(x[1:3], y[1:3], color = 'pink', alpha=0.5) #alpha : 투명정도
# 네점을 잇는 영역에 색을 채운다.
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)

plt.fill_between(x[1:3], y1[1:3], y2[1:3], color = 'lightgray', alpha = 0.5)

plt.show()
```

```python
t = np.arange(0, 2, 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bo', t, t**3, 'g-.')
plt.show()
```

```python
t = np.arange(0, 2, 0.2)

plt.plot(t, t, 'bo')                              # 선의 두께
plt.plot(t, t**2, color = '#e35f62', marker='*', linewidth = 2)
plt.plot(t, t**3, color = 'springgreen', marker='^', markersize = 9)
plt.show()
```

```python
t = np.arange(0, 2, 0.2)

plt.plot(t, t, 'bo')                              # 선의 두께
plt.plot(t, t**2, color = '#e35f62', marker='*', linewidth = 2)
plt.plot(t, t**3, color = 'springgreen', marker='^', markersize = 9)
plt.grid(True)
plt.show()
```

```python
# 가로방향의 그리드 표시
# Grid(격자, 눈금자)
t = np.arange(0, 2, 0.2)

plt.plot(t, t, 'bo')                              # 선의 두께
plt.plot(t, t**2, color = '#e35f62', marker='*', linewidth = 2)
plt.plot(t, t**3, color = 'springgreen', marker='^', markersize = 9)
# axis : both(기본, default), x, y
plt.grid(True, axis = 'y')
plt.show()
```

```python

t = np.arange(0, 2, 0.2)

plt.plot(t, t, 'bo')                              
plt.plot(t, t**2, color = '#e35f62', marker='*', linewidth = 2)
plt.plot(t, t**3, color = 'springgreen', marker='^', markersize = 9)
# axis : both(기본, default), x, y
plt.grid(True, axis = 'y', color = 'red', alpha = 0.5, linestyle = '--')
plt.show()
```

```python
# 틱(Tick) : 축의 간격을 표시하는 눈금
# xticks() - x축, yticks() - y축
t = np.arange(0, 2, 0.2)

plt.plot(t, t, 'bo')                              
plt.plot(t, t**2, color = '#e35f62', marker='*', linewidth = 2)
plt.plot(t, t**3, color = 'springgreen', marker='^', markersize = 9)
plt.xticks([0, 1, 2])
plt.yticks(np.arange(1, 6))
plt.show()
```

```python
# 눈금 레이블
t = np.arange(0, 2, 0.2)

plt.plot(t, t, 'bo')                              
plt.plot(t, t**2, color = '#e35f62', marker='*', linewidth = 2)
plt.plot(t, t**3, color = 'springgreen', marker='^', markersize = 9) # 담겨있을 공간이 없으면 지워짐
plt.xticks(np.arange(0, 2, 0.2), labels = ['Jan',' ', 'Feb',' ', 'Mar',' ','May',' ', 'June',' ', 'July'])
plt.yticks(np.arange(0, 7), ('0','1GB','2GB', '3GB', '4GB', '5GB', '6GB'))
plt.show()                 # lables 대신 () 형태로 대체가능.
```
