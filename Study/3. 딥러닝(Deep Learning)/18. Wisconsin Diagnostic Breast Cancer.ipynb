### 데이터 로딩하기

```python
breast = pd.read_csv('D:/python3/data/data.csv')
breast.head()
```

```python
breast.info()
```

- target : diagnosis, M(malignant) = 1(악성종양 = 암), B(benign) = 0

```python
breast.isnull().sum()
```

```python
breast['diagnosis'].unique()
```

```python
def func(row):
    if 'M' in row:
        return 1
    else:
        return 0

# target라는 변수 만들기
breast['target'] = breast['diagnosis'].apply(func)
breast.head()
```

```python
cancer = breast.drop(columns = ['id', 'diagnosis', 'Unnamed: 32'])
cancer.head()
```

### train/test data set 분할

```python
X = cancer.drop(['target'], axis = 1)
y = cancer['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state = 42,
                                                   stratify = y)
```

```python
# 앙상블 : 하나의 데이터를 여러개의 분류기를 통해 다수의 학습 모델을 만들어 학습시키고 
#         학습 결과를 결합함으로써 과적합을 방지하고 정확도를 높이는 학습 기법 

# 보팅: 앙상블 학습의 기본, 여러 개의 분류기가 투표를 통해 최종 예측 결과를 결정하는 방식
#       서로 다른 알고리즘을 여러 개 결합하여 사용

#      하드 보팅(Hard Voting)
#          다수의 분류기가 예측한 결과값을 최종 결과로 선정
#      소프트 보팅(Soft Voting)
#          모든 분류기가 예측한 레이블 값의 결정 확률 평균을 구한 뒤 가장 확률이 높은 
#          레이블 값을 최종 결과로 선정

# 배깅: 하나의 데이터를 여러개로 나누어 학습하는 앙상블 학습법

#      보팅을 하나의 고유한 학습법으로 보고 배깅과 구분하자면, 
#      일반적으로 보팅은 하나의 데이터에 여러 알고리즘 적용, 
#      배깅은 여러개로 나누어진 데이터에 하나의 알고리즘을 적용하는 것으로 구분
#      하지만 여러개로 나누어진 데이터를 이용하는 배깅에서도, 
#      최종 예측값을 선택하는 행위는 '보팅'이라 함

# 부스팅: 여러개의 분류기가 순차적으로 학습을 수행
#        이전 분류기가 예측이 틀린 데이터에 대해서 올바르게 예측할 수 있도록 
#        다음 분류기에게 가중치(weight)를 부여하면서 학습과 예측을 진행
#        계속하여 분류기에게 가중치를 부스팅하며 학습을 진행
# 랜덤 포레스트: 배깅 + 의사결정 나무
```

### 투표(voting)기법

```python
logistic = LogisticRegression(C = 0.001, random_state = 42)
tree = DecisionTreeClassifier(max_depth = None, criterion = 'entropy', random_state = 42)
knn = KNeighborsClassifier(n_neighbors = 1, p= 2, metric = 'minkowski')
# metric : 나무에 사용할 거리 측정법입니다. 
#         기본 메트릭은 minkowski이고 p = 2 인 경우 표준 유클리드 메트릭과 동일합니다.

voting_estimators = [('logistic', logistic), ('tree', tree), ('knn', knn)]

voting = VotingClassifier(estimators = voting_estimators, voting = 'soft')
# hard 다수결 / soft 확률높은것

clf_labels = ['logistic regression', 'Decision tree', 'KNN', 'Majority voting']
all_clf = [logistic, tree, knn, voting]
```

```python
voting.fit(X_train, y_train)
pred = voting.predict(X_test)
print('보팅 분류기의 정확도 :{0:.4f}'.format(accuracy_score(y_test, pred)))
```

```python
# 개별 모델의 학습/예측/평가
classifiers = [logistic, tree, knn]
for i in classifiers:
    i.fit(X_train, y_train)
    pred = i.predict(X_test)
    class_name = i.__class__.__name__
    print('{0}의 정확도 :{1:.4f}'.format(class_name, accuracy_score(y_test, pred)))
```
