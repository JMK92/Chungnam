### k-mean 클러스터링
- 분류(Classifier)-(지도학습)와 군집(Clustering)-(비지도학습)은 다르다.
- 비지도 학습(Unsupervised Learning) 알고리즘
- k개로 cluster(군집) == class
- mean : 각 클러스터의 중심과 데이터들의 평균거리

#### k-mean 클러스터링 과정
- step1. 데이터
- step2. 몇개의 클러스터를 만들지 결정
- step3. 클러스터링할 때, 클러스터의 중심(centeroid)을 설정

```python
# 군집은 정답이 없다. 
# 데이터 중심으로 각 군집과 거리를 계산해서 가까원 군집으로 속해 진다.
# 유클리드 사용하여 중심을 찾는다.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import random
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
# 데이터 생성
df = pd.DataFrame(columns = ['x', 'y'])
```

```python
for i in range(30):
    df.loc[i] = [random.randrange(1, 100), 
                 random.randrange(1, 100)]
```

```python
df
```

```python
# 시각화
import seaborn as sns
sns.lmplot(x = 'x', y = 'y', data = df, 
           fit_reg = False, scatter_kws = {'s':30})
plt.title('K-means')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 데이터를 numpy 배열로 
points = df.values
```

```python
# k-mean 알고리즘 3개의 cluster를 생성
kmeans = KMeans(n_clusters = 3).fit(points)
```

```python
# 중심값 생성
kmeans.cluster_centers_
```

```python
# 데이터가 포함된 클러스터
kmeans.labels_
```

```python
df['cluster'] = kmeans.labels_
df
```

```python
import seaborn as sns
sns.lmplot(x = 'x', y = 'y', data = df, 
           fit_reg = False, scatter_kws = {'s':100}, hue = 'cluster')
plt.title('K-means')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
