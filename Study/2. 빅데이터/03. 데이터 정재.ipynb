```python
df.index
```

### 결측치(NaN)

```python
df['E'] = [1.0, np.nan, 3.5, 6.1, np.nan, 7.0]
df
```

```python
df.dropna(how = 'all') # 결측치 제거
```

```python
df.dropna(how = 'any') # how = 'any' : 어느 하나라도 결측치 있으면 제거
```

```python
# fillna(value = 5.0) # 5.0으로 결측치 대신에 채워라 or value 빼고 5.0만 적어도 같은 의미
df
```

```python
df.fillna(5.0)
```

```python
# isnull() : 결측값입니까? True or False
df.isnull()
```

```python
# 결측값이 있는 row 값만 빼오기
df.loc[df.isnull()['E'], :]
```
