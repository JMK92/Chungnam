```python
# up % down game
import numpy as np

num = np.random.choice(50, 1, replace=False)

while True:
    select = int(input('수 = '))
    if select > num:
        print('DOWN')
    elif select < num:
        print('UP')
    else:
        print('정답')
        break
        
    if select == 0:
        break
```
