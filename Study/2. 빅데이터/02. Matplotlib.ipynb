### 산점도(산포도) : plot(x, y)
- scatter()

```python
data = []
for i in range(100):
    x = np.random.normal(0, 1) # normal -> 정규분포에서 추출해라
    y = x *0.1 +0.2 + np.random.normal(0, 1)
    data.append([x, y])
```

```python
x_data = [v[0] for v in data]
#x_data
```

```python
y_data = [v[1] for v in data]
#y_data
```

```python
plt.plot(x_data, y_data, 'ro') # red marker => 'ro'
plt.show()
```

### 화면에 여러개의 그래프를 그리기
- figure()
- add_subplot()

```python
# figure 객체
# 하나의 화면에 한개의 그래프가 있을 때는 figure 생략
# 하나의 화면에 여러개의 그래프를 그릴때는 figure 넣어야함.
fig = plt.figure()
# 2 x 2 열로 나누어 4개의 그래프를 그린다.
x1 = fig.add_subplot(2, 2, 1) # 2행 2열의 첫번째 영역.
x2 = fig.add_subplot(2, 2, 2)
x3 = fig.add_subplot(2, 2, 3)
x4 = fig.add_subplot(2, 2, 4)

# data 생성
# 1 ~ 100 숫자중 난수 정수를 발생, 100개
data = np.random.randint(1, 100, 100) # 정규분포면서 정수

# 첫번째 영역
x1.hist(data)

# 두번째 영역
x2.scatter(x_data, y_data)

# 세번째 영역
x3.plot(np.arange(50))

# 네번째 영역
# 'b' : blue, 'k' : black
x4.plot(np.random.randn(100), 'k--')

plt.show()
```

### 하나의 차트에 여러개의 그래프를 담는 작업.

```python
fig = plt.figure()
graph = fig.add_subplot(1, 1, 1) # (행, 열, 몇번째)

x = np.random.randn(50) # 정규분포난수
y = np.random.randn(50).cumsum()

# 계단형 그래프                             step 형태 라는 뜻
graph.plot(x, color = 'r', label = 'step', drawstyle = 'steps-post')

# 선스타일, # 선스타일을 바꾸려면  color를 빼야함.
graph.plot(y, 'g--', label = 'line') # line은 default임.

# 범례    최적의 장소에 범례
plt.legend(loc=0) # 오른쪽 상단, - 'upper right' or 'loc = 1' 
                    # 왼쪽 상단 - 'loc = 2', 왼쪽 하단 - 'loc = 3' 
                   # 최적의 장소 : loc='best'
# 차트 제목 
plt.title('multi chart plotting')

# 축 레이블
plt.xlabel('stage')
plt.ylabel('random number')

plt.show()
```

```python
fig = plt.figure()
 
ax = fig.add_subplot(1, 1, 1) # fig 객체 만들면 fig.add_subplot 해줘야함.
plt.plot(np.random.randn(1000).cumsum()) # cumsum() : 누적합

# set_xticks는 fig 객체를 만들어야 함.
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
                           rotation = 30)
plt.show()
```

### iris 데이터셋

```python
iris = pd.read_csv('D:/python3/data/iris.csv')
iris.head(10)
```

```python
iris.tail(10)
```

```python
iris.info()
```

```python
# Sepal.Length와 Petal.Length를 이용해 산점도 그래프
plt.scatter(iris['Sepal.Length'], iris['Petal.Length'])
plt.xlabel('Sepal.Length')
plt.ylabel('Petal.Length')
plt.title('Iris Dataset')
plt.show()
```

```python
# 산점도 행렬 : scatter_matrix()
iris_col = iris.loc[:, 'Sepal.Length':'Petal.Width']
pd.plotting.scatter_matrix(iris_col, diagonal= 'kde')
plt.show()
```

```python
iris['Species'].unique()
```

```python
cols = []
for s in iris['Species']:
    ch = 0
    if s == 'setosa':
        ch = 1
    elif s == 'versicolor':
        ch = 2
    else:
        ch = 3
    cols.append(ch)
    
plt.scatter(iris['Sepal.Length'], 
            iris['Petal.Length'],
           c = cols) # c : colors
plt.xlabel('Sepal.Length')
plt.ylabel('Petal.Length')
plt.title('Iris Dataset')
plt.show()
```
