```python
# 파생변수 : 5대 범죄에 검거율을 계산하여 각 컬럼에 대입
crime_anal_raw['강간검거율'] = crime_anal_raw['강간 검거'] / crime_anal_raw['강간 발생']*100
crime_anal_raw['강도검거율'] = crime_anal_raw['강도 검거'] / crime_anal_raw['강도 발생']*100
crime_anal_raw['살인검거율'] = crime_anal_raw['살인 검거'] / crime_anal_raw['살인 발생']*100
crime_anal_raw['절도검거율'] = crime_anal_raw['절도 검거'] / crime_anal_raw['절도 발생']*100
crime_anal_raw['폭력검거율'] = crime_anal_raw['폭력 검거'] / crime_anal_raw['폭력 발생']*100

# 불필요한 컬럼 제거
del crime_anal_raw['강간 검거']
del crime_anal_raw['강도 검거']
del crime_anal_raw['살인 검거']
del crime_anal_raw['절도 검거']
del crime_anal_raw['폭력 검거']

crime_anal_raw.head()
```

```python
# 퍼센트가 100이 넘는 것을 100으로 처리
col_list = ['강간검거율', '강도검거율', '살인검거율', 
            '절도검거율', '폭력검거율']

for c in col_list:
    crime_anal_raw.loc[crime_anal_raw[c] > 100, c] = 100

crime_anal_raw.head()
```

```python
# 피처(특성, 컬럼) 명 변경
crime_anal_raw.rename(columns= {'강간 발생':'강간',
                               '강도 발생': '강도',
                               '살인 발생': '살인',
                               '절도 발생': '절도',
                               '폭력 발생': '폭력'}, inplace=True)
crime_anal_raw.head()
```

### 정규화(Normalization)
- 데이터의 최소 0, 최대 1 사이의 비율값을 대체
- x_new = x - x_min / x_max - x_min : 숫자 데이터의 정규화 공식 -> 파이썬에 만들어져 있음
- from sklearn.preprocessing import MinMaxScaler
- scaler = MinMaxScaler()
- df[:] = scaler.fit_transform(df[:]) - 단한번으로 0과 1사이로 변환이 됨.

```python
from sklearn import preprocessing
#-------------------- 하는 이유 ----------------------------
# 시각화 할려고 하는데 숫자의 차이 너무 나서 시각화에서 안 보일 수 있기 때문에
# 정규화하여 비슷하게 맞춰 시각화에서 잘 보일 수 있게 한다.
# ------------------------------------------------------------

# 정규화 대상 컬럼만 추출
col = ['강간', '강도', '살인', '절도', '폭력']

x = crime_anal_raw[col].values
min_max_scaler = preprocessing.MinMaxScaler() # 객체 만들기
x_scaled = min_max_scaler.fit_transform(x.astype(float))

crime_anal_norm = pd.DataFrame(x_scaled, columns = col, 
                               index = crime_anal_raw.index)
#---------------------- col 만 가지고 있음 --------------------------
col2 = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']

# --------------------- col2  추가  -----------------------------------
crime_anal_norm[col2] = crime_anal_raw[col2]

crime_anal_norm.head()
```

```python
# 서울시 cctv 현황 분석의 결과 파일을 불러와 구별로 가져온 자료중 
# 인구수와 소계를 범죄 현황 데이터에 포함시키는 작업
result_CCTV = pd.read_csv('D:/python3/data/CCTV_Result.csv',
                         encoding='utf-8',
                         index_col = '구별')
result_CCTV
```

```python
crime_anal_norm[['인구수', 'CCTV']] = result_CCTV[['인구수', '소계']]
crime_anal_norm.head()
```

```python
# 시각화를 하기 위하여
# 각 범죄의 발생 건수를 범죄라는 이름으로 합하는 작업
col = ['강간', '강도', '살인', '절도', '폭력']
crime_anal_norm['범죄'] = np.sum(crime_anal_norm[col], axis=1)
crime_anal_norm.head()
```

```python
# 각 범죄의 검거율을 검거라는 이름으로 합하는 작업
col2 = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
crime_anal_norm['검거'] = np.sum(crime_anal_norm[col2], axis=1)
crime_anal_norm.head()
```

### 지도시각화(Folium-포리움)
- 파이썬에서 제공하는 라이브러리로 지도를 다루는 대표적인 라이브러리
- 외부 : pip install folium
- 로딩 : import folium as g
- 위도와 경도를 이용해서 시각화
    1. Map객체             zoom_start : 처음 나오는 지도 배율
           Map(location[위도, 경도], zoom_start= 10)

```python
# 외부에서 anaconda prompt shell 관리자 권하으로 실행하여 pip install folium 해줌
import folium as g
```

```python
# Map객체을 사용해서 location 옵션에 위도, 경도를 설정
# 연돈 : 33.25909, 126.40727
g.map = g.Map(location = [33.25909, 126.40727], zoom_start=18)
g.map
# 이렇게 보면 어디가 연돈 인지 모른다 그래서 다음장에 마커 표시
```

```python
# marker(표시)
marker = g.Marker([33.25909, 126.40727], popup= '연돈',
                  icon= g.Icon(color = 'blue'))
marker.add_to(g.map)
g.map
```

```python
# 범위를 원형태로 지정하는 작업,  CircleMarker
marker2 = g.CircleMarker([33.25909, 126.40727],
                       radius = 100,         #원의 범위
                       color = 'skyblue',    #원의 실선 색
                       fill_color = 'skyblue', # 원 채우는 색
                       popup = '연돈')
marker2.add_to(g.map)
g.map
```
